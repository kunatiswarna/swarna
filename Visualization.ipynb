```python
from ggplot import *
import pandas as pd
%matplotlib inline
```

```python
mpg = ggplot.mpg
```

```python
mpg.head()
```

### Univariate

```python
mpg.cty.hist()
```

```python
mpg.manufacturer.value_counts().plot(kind='bar')
```

```python
mpg.boxplot('cty')
```

```python
mpg.groupby(['year','cyl'])['cty'].mean().plot(kind='bar')
```

## GGplot

```python
ggplot(mpg,aes('cty','hwy'))+geom_point()
```

```python
ggplot(mpg,aes('cty','displ'))+geom_abline()
```

```python
ggplot(mpg,aes('cty','hwy',color='class'))+geom_point()
```

```python
ggplot(mpg,aes('cty','hwy',color='class',shape='drv'))+geom_point()
```

```python
ggplot(mpg,aes('cty','hwy',size='class',shape='drv'))+geom_point()
```

```python
x = ggplot(mpg,aes('cty','hwy',color='class',shape='drv'))+geom_point(size=20)
```

```python
x + facet_grid('fl')
```

```python
x + facet_wrap('fl','year',)+theme_bw()
```

```python
days = list(range(0, 22, 3))
celsius_values = [25.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
plt.plot(days, celsius_values,"-",
        days, celsius_values,"o")
plt.show()
```

```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.plot(days,celsius_min,"o",
         days,celsius_min,"-",
        days,celsius_max,"o",
        days,celsius_max,"-")
plt.xlabel('Day')
plt.ylabel('Degrees Celsius')
plt.show()
```

```python
days = list(range(1,9))
celsius_min = [19.6, 24.1, 26.7, 28.3, 27.5, 30.5, 32.8, 33.1]
celsius_max = [24.8, 28.9, 31.3, 33.0, 34.9, 35.6, 38.4, 39.2]
plt.scatter(celsius_min,celsius_max,c=days)
```

```python
years = (2010, 2011, 2012, 2013, 2014)
menMeans = (20, 35, 30, 35, 27)
womenMeans = (25, 32, 34, 20, 25)
plt.bar(years,menMeans,color='r')
plt.bar(years,womenMeans,color='b')

plt.ylabel('Scores')
plt.xlabel('Years')
plt.title('Scores by group and gender')
```
